```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm

%matplotlib inline
```

```python
# importing data
df = pd.read_excel('/mnt/d/Desktop/Marketing_Analytics/Case Assignment 1 Data - Vodka Data.xlsx', 
                     sheetname = 'data')
df.head()
```

### Problem 1

```python
lm_1 = sm.ols(formula='LnSales ~ PriceRerUnit + Print + Outdoor + Broad + LagTotalSales', data = df).fit()
lm_1.summary()
```

### Problem 2

```python
lm_2 = sm.ols(formula='LnSales ~ LnLPrice + PriceRerUnit + LnPrint + LnOut + LnBroad', 
              data = df).fit()
lm_2.summary()
```

### Problem 3

```python
lm_3 = sm.ols(formula='LnSales ~ LnLPrice + PriceRerUnit + LnPrint + LnOut + LnBroad + Tier1 + Tier2', 
              data = df).fit()
lm_3.summary()
```

### Problem 4

```python
lm_4 = sm.ols(formula='LnSales ~ LnLPrice + PriceRerUnit + LnPrint + LnOut + LnBroad + Tier1 + Tier2 + \
              LagTotalMinusSales', data = df).fit()
lm_4.summary()
```
